## Description
This project is a  classification project aimed at predicting user preferences for mountains or beaches based on various features like Annual_Income, Budget,EDucation_Level etc,in the dataset. It utilizes key concepts of  __Python programming, machine learning, data preprocessing, feature engineering__, and __model evaluation__ to create a functional, efficient, and user-friendly solution.

- __Activity Preference__: Categorical data (e.g., adventure, relaxation, sightseeing, sports, cultural, nature).
- __Climate Preference__: Categorical data (e.g., cold, warm, tropical, temperate, dry, wet).
- __Travel Frequency__: Numerical data (e.g., 0–12 trips per year).
- __Accommodation Type__: Categorical data (e.g., hotel, resort, hostel, vacation rental).
- __Budget__: Numerical data (e.g., range from $100 to $10,000).
- __Group Size__: Numerical data (e.g., 1 for solo, 2–4 for small groups, 5+ for larger groups).
- __Preferred Destination__: Categorical data (e.g., beach, mountains, cities, countryside, international).
- __Trip Duration__: Numerical data (e.g., 1–7 days, weekend, one week, longer trips).
- __Mode of Transportation__: Categorical data (e.g., car, flight, train, bus).
- __Accommodation Rating__: Categorical data (e.g., 1–5 stars, basic, mid-range, luxury).
- __Travel Purpose__: Categorical data (e.g., leisure, business, education, family visit).
- __Health Considerations__: Categorical data (e.g., medical needs, accessibility, dietary restrictions).
- __Environmental Impact Concern__: Categorical data (e.g., low, medium, high concern about sustainability).

## Data Preprocessing

### Handling Missing Values

### Data Scaling
- **Min-Max Normalization** is applied to the following columns:
  - `Age`
  - `Income`
  - `Vacation_Budget`

- **Z-Score Normalization** is applied to the following columns:
  - `Proximity_to_Mountains`
  - `Proximity_to_Beaches`
  - `Travel_Frequency`

### Data Encoding
- **One-Hot Encoding** is applied to the following categorical columns:
  - `Gender`
  - `Preferred_Activities`
  - `Location`

- **Label Encoding** is applied to the following column:
  - `Education_Level`

- **Ordinal Encoding** is applied to the following column:
  - `Favorite_Season`
